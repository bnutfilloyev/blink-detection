```python
from helpers import *
import matplotlib.pyplot as plt
import os, glob, cv2, random
import seaborn as sns
import pandas as pd
```

```python
base_path = 'dataset.csv'
X, y = read_csv(base_path)

print(X.shape, y.shape)
```

```python
plt.figure(figsize=(15, 15))
for i in range(100):
    plt.subplot(10, 10, i+1)
    plt.axis('off')
    plt.imshow(X[i].reshape((26, 34)), cmap='gray')
```

```python
sns.distplot(y, kde=False)
```

```python
n_total = len(X)
X_result = np.empty((n_total, 26, 34, 1))

for i, x in enumerate(X):
    img = x.reshape((26, 34, 1))
    X_result[i] = img
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(X_result, y, test_size=0.1)

print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)

np.save('datasets/x_train.npy', x_train)
np.save('datasets/y_train.npy', y_train)
np.save('datasets/x_val.npy', x_val)
np.save('datasets/y_val.npy', y_val)
```
